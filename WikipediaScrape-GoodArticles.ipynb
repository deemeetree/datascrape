```python
import wikipediaapi
import requests
from bs4 import BeautifulSoup
import time
import numpy as np
from random import randint


import xml.etree.ElementTree as ET

# create the file structure for articles
root = ET.Element('articles')

# create the file structure for ground truth
rootgt = ET.Element('articles')

# prettify XML
def indent(elem, level=0):
    i = "\n" + level*"  "
    if len(elem):
        if not elem.text or not elem.text.strip():
            elem.text = i + "  "
        if not elem.tail or not elem.tail.strip():
            elem.tail = i
        for elem in elem:
            indent(elem, level+1)
        if not elem.tail or not elem.tail.strip():
            elem.tail = i
    else:
        if level and (not elem.tail or not elem.tail.strip()):
            elem.tail = i

#print(ET.tostring(head).decode())

parse_url = 'https://en.wikipedia.org/wiki/Wikipedia:Good_articles/all'

start_item = ''

def parse_wiki(start_url, start_item, iterations, lastcount):
    
    runtimes = 0
    
    # first pull the HTML from the page that links to all of the pages with the links.
    html = requests.get(start_url + start_item + '')

    #turn the HTML into a beautiful soup text object
    b = BeautifulSoup(html.text, 'lxml')
    # create an empty list where those links will go.
    links = []

    # in this case, all of the links we're in a '<li>' brackets.
    for divs in b.findAll("div", {"class": "NavContent"}):
        for pars in divs.find_all('p'):
            # pull the actual link for each one
            for link in pars.find_all('a', href=True):
                # we don't want all the articles, only 600 from ~30000, so we introduce this random element
                dice = randint(1, 50)
                if dice == 50:
                    links.append(link['href'])
    # the above code ends up pulling more links than I want,
    # so I just use the ones I want
    
    start_from = 0
    finish_at = len(links)
    total_processed = finish_at - start_from
    
    links = links[start_from:finish_at] #29:227
    
    # remove "wiki" from the link
    full_links = [i[6:] for i in links]

    wk = wikipediaapi.Wikipedia('en',extract_format=wikipediaapi.ExtractFormat.WIKI)

    count = 0
    last_title = ''

    for link in full_links:
     
        page = wk.page(link)    
        last_title = page.title

        if (count < total_processed):
            article = ET.SubElement(root, 'article', {'id':str(lastcount),'title':page.title})
            articlegt = ET.SubElement(rootgt, 'article', {'id':str(lastcount),'npov':'true','title':page.title})
            article.text = page.text
            print(page.title)
            lastcount = lastcount + 1
            count = count + 1
        
    indent(root)
    indent(rootgt)
    
    
parse_wiki(parse_url, start_item, 1, 792)

tree = ET.ElementTree(root) 
treegt = ET.ElementTree(rootgt) 

tree.write('article-training-wikipedia-short-NPOV.xml', encoding='utf-8', xml_declaration=True)
treegt.write('ground-truth-wikipedia-short-NPOV.xml', encoding='utf-8', xml_declaration=True)




    
```
